```python
import glob
import os
import cv2
import json
import io

coco_val = ["truck", "car", "bus", "person", "rider", "bicycle", "caravan", "motorcycle", "train", "trailer"]
# coco_train = ["car", "person", "truck", "bus", "bicycle", "rider", "trailer", "motorcycle", "caravan", "train"]

label = dict()
for idx,item in enumerate(coco_val):
    label.update({idx : item})

labelme_json_img_path = "G:/Bdd100k/labelme/val"
coco_json_path = "G:/Bdd100k/outputs/labels/bdd100k_val.json"
jpg_path = "E:/instance segmentation dataset/bdd100k_ins_seg_labels_trainval/bdd100k/imgs/val"   

file_json = io.open(coco_json_path, 'r', encoding='utf-8')
m_json_data = file_json.read()
m_data = json.loads(m_json_data)

for item in m_data['images']:
    flag=False
    m_images_file_name = item['file_name']
    print(m_images_file_name)
    
    (m_name, extension) = os.path.splitext(m_images_file_name)
   
    m_image = cv2.imread(os.path.join(jpg_path, m_images_file_name))
    m_images_height = m_image.shape[0]
    m_images_width = m_image.shape[1]
    m_images_id = item['id']
    
    data = {}
    data['version'] = "5.1.1"
    data['flags'] = {}
    data["shapes"] = []
    
    for idx in range(len(m_data['annotations'])):
        annit = m_data['annotations'][idx]
        m_image_id = annit['image_id']
        m_category_id = annit['category_id']
        
        if m_image_id == m_images_id:
            flag = True 
            points = []
            for segitem in annit['segmentation']:
                points.append(segitem)
                
            itemData = {'points': []}
            itemData["label"] = label[m_category_id]
            itemData['points'].extend(points)
            itemData["group_id"] = None
            itemData["shape_type"] = "polygon"
            itemData["flag"] = {}
            data["shapes"].append(itemData)
            
    if flag:
        data['imagePath'] = m_images_file_name
        data['imageData'] = None
        data['imageHeight'] = m_images_height
        data['imageWidth'] = m_images_width
        
        jsonName = ".".join([m_name, "json"])
        jpgName = ".".join([m_name, "jpg"])
        jsonPath = os.path.join(labelme_json_img_path, jsonName)
        jpgPath = os.path.join(labelme_json_img_path, jpgName)
        
        with open(jsonPath, "w") as f:
            json.dump(data, f, indent = 2)
        cv2.imwrite(jpgPath, m_image)
        

print('{} files'.format(len(os.listdir(labelme_json_img_path))))
print("File processing completed...")
```
